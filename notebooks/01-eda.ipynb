# Selecting with Location-Based Indexing

This is the recommended way to select or subset data from a DataFrame in `pandas`

`df.iloc`: Purely integer-location based indexing for selection by **position**.

`df.loc`: Access a group of rows and columns by **label(s)** or a boolean array.

Arguments:

`(row, column)`
`(start:stop:step, start:stop:step)`

Examples:

## iloc

```
iloc[rows,columns]
iloc[:,:]
iloc[start:stop:step, start:stop:step] 
# NB: iloc is inclusive for start, exclusive for stop
```

## loc

```
loc[rows, columns]
loc[:,:]
loc[start:stop:step, start:stop:step]
loc[:, [1, 2, 3]]
# NB: loc is inclusive for start, inclusive for stop
```

# `loc` vs `iloc`

````python
import random
````

````python
# Randomly scramble the index of the DataFrame
random_index = random.sample(list(wine.index), k = len(list(wine.index)))
# Create a new column with this random index
wine.loc[:, 'scrambled_index'] = random_index
# Change the index of the dataset to the scrambled index
scrambled_wine = wine.set_index('scrambled_index')
scrambled_wine
````

````python
# Select first 5 rows
scrambled_wine.head(5)
````

````python
# Select first 5 rows with iloc (integer indexing)
scrambled_wine.iloc[0:5,:]
````

````python
# loc uses value rather than integer indexing
# returns all rows up to where #5 is in the index
# #5 is no longer in the 5th position
# because we scrambled the index
# #5 is at position 76954
scrambled_wine.loc[:5]
````

````python
# we can search for index location of the #5 using get_loc 
scrambled_wine.index.get_loc(5)
````

````python
# iloc uses integer indexing and so locates the 5th row
scrambled_wine.iloc[:5]
````

````python
# Select rows 0-5 (exclusive for stop) and columns 0-5 (exclusive for stop) using iloc
scrambled_wine.iloc[:5,:5]
````

````python
# Select first 5 rows and columns "points" and "region_1" of the scrambled_wine table
# loc first
scrambled_wine.loc[:,["points","region_1"]].iloc[0:5]
````

````python
# Select first 5 rows and columns "points" and "region_1" of the scrambled_wine table
# iloc first
scrambled_wine.iloc[0:5].loc[:,["points","region_1"]]
````

````python
# Select first 5 rows of points column
scrambled_wine.loc[:,"points"].iloc[0:5,]
````

````python
# Select first 5 rows of region_1 column
scrambled_wine.loc[:,"region_1"].iloc[0:5,]
````

````python
# Sort the index of the scrambled_wine DataFrame
scrambled_wine.sort_index()
````

````python
# Reset the index of the scrambled_wine DataFrame
# Note that the scrambled_index column is kept
# The DataFrame is reordered by row numbers, and moves the previous index to a column variable
scrambled_wine.reset_index()
````

````python
# Let's revert to the previous scrambled_wine DataFrame
# Randomly scramble the index of the DataFrame
random_index = random.sample(list(wine.index), k = len(list(wine.index)))
# Create a new column with this random index
wine.loc[:, 'scrambled_index'] = random_index
# Change the index of the dataset to the scrambled index
scrambled_wine = wine.set_index('scrambled_index')
scrambled_wine
````

````python
# Sort the DataFrame and assign it to a new DataFrame variable
sorted_wine = scrambled_wine.sort_index()
sorted_wine
````
